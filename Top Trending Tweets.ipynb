```python
import socket
import requests
import requests_oauthlib
import json
import pickle
import numpy as np
import pandas as pd

# Replace the values below with yours
ACCESS_TOKEN = ''
ACCESS_SECRET = ''
CONSUMER_KEY = ''
CONSUMER_SECRET = ''
my_auth = requests_oauthlib.OAuth1(CONSUMER_KEY, CONSUMER_SECRET,ACCESS_TOKEN, ACCESS_SECRET)

def top_tweets():
    dict_woeid={23424908:"Nigeria",23424977:"United States",23424856:"Japan",23424975:"United Kingdom",23424938:"Saudi Arabia",23424969:"Turkey",23424768:"Brazil",23424848:"India",23424900:"Mexico",23424950:"Spain",23424819:"France"}
    data={}
    for key in dict_woeid.keys():
        url = 'https://api.twitter.com/1.1/trends/place.json?id='+str(key)
        response = requests.get(url, auth=my_auth, stream=True)
        trend=response.json()[0]["trends"]
        tweets=[] 
        tweets.append("Hashtag : Tweet Volume")
        tweets_top=[]
        for x in trend:
            if x['tweet_volume']!=None:
                tweets_top.append((x['name'],x['tweet_volume']))
        tweets_top.sort(key= lambda x:x[1],reverse=True)
        tweets_top=tweets_top[:5]
        for x in tweets_top:
            tweets.append(x[0]+" : "+str(x[1]))
        data["Country: "+dict_woeid[key]]=tweets
    return pd.DataFrame(data)

```
